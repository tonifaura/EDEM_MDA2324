```python
!pip install unidecode
!pip install nltk
import matplotlib.pyplot as plt
from collections import Counter
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

```

```python
#Abriendo el archivo de texto proporcionado por el .py

file = open ('value.txt', 'r')
texto = file.read()
texto
```

```python
# Pasar texto a lower cases

texto_low = texto.lower()
texto_low
```

```python
# Dividiendo el texto en listas
# Lista con palabras como elementos en minuscula

texto_list = texto_low.split()
```

```python
# Quitando las "Stop Words" para obtener un analisis mas limpio

stop_words = set(stopwords.words('english')) #para obtener las stop words

lista_sinsw = [palabra for palabra in texto_list if palabra.lower() not in stop_words]

lista_sinsw
```

```python
# Lista que contiene palabras sin comas

lista_sincoma = [palabra.rstrip(",") for palabra in lista_sinsw]

```

```python
# Lista que contiene palabras sin puntos

lista_sinpunto = [palabra.rstrip(".") for palabra in lista_sincoma]
```

```python
# Lista que contiene palabras sin espacios al final

lista_sinespacios = [palabra.rstrip() for palabra in lista_sinpunto]
```

```python
# Listas que contiene palabras sin apostrofes (') al final

lista_sinapostrofe = [palabra.rstrip("'") for palabra in lista_sinespacios]
```

```python
# Listas que contiene palabra sin signo de pregunta (?) al final

lista_sinpregunta = [palabra.rstrip("?") for palabra in lista_sinapostrofe]
```

```python
# Listas que contiene palabra sin signo de exclamacion (!) al final

lista_sinexclam = [palabra.rstrip("!") for palabra in lista_sinpregunta]
```

```python
# Listas que contiene palabra los siguientes simbolos ( ":) )  al final (cled=clean end)

lista_clend = [palabra.rstrip(' ":) ') for palabra in lista_sinexclam]
```

```python
# Listas que contiene palabra los siguientes simbolos ( (" )  al principio (clebeg=clean begenning)

lista_clebeg = [palabra.lstrip(' (" ') for palabra in lista_clend]
```

```python
# Contando la frecuencia que tiene cada palabra

contador_palabra = Counter(lista_clebeg)
```

```python
# Obteniendo las palabras y sus frecuencias

palabras = list(contador_palabra.keys())
frecuencias = list(contador_palabra.values())
```

```python
# Graficando el conteo de palabras

plt.bar(palabras, frecuencias)
plt.xlabel('palabras')
plt.ylabel('frecuencias')
plt.title('Conteo de palabras repetidas de Chuck')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# Mostrando solo las 20 palabras mas repetidas

palabra_top = [palabra[0] for palabra in contador_palabra.most_common(20)]
frecuencia_top = [contador_palabra[palabra] for palabra in palabra_top ]

plt.bar(palabra_top, frecuencia_top)

plt.xlabel('palabras')
plt.ylabel('frecuencias')
plt.title('Fr de las 20 palabras mas repetidas en los chistes de Chuck')
plt.xticks(rotation=45, ha='right')

plt.show()

```

```python
# Graficando en grafica circular

plt.pie(frecuencia_top, labels=palabra_top, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Hace que el grfico sea un crculo en lugar de una elipse
plt.title('Top 20 Palabras Ms Repetidas')
plt.show()
```
