# 1.- Importo las librerías

```python
import requests
import json
import time
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

# 2.- Codigo para traer las frases de Chuck de la URL:

```python
url = "https://api.chucknorris.io/jokes/random"

numeroFrases=0

while numeroFrases<=10:

    time.sleep(1)
    respuesta = requests.get(url)
    datos=respuesta.json()
    fraseChuck:str =datos["value"]
    numeroFrases+=1
    f = open("/content/drive/MyDrive/Colab Notebooks/EjercicioRepasoClase091123.txt", "a")
    f.write( str(numeroFrases) +" " + fraseChuck + "\n")
    f.close()

f = open("/content/drive/MyDrive/Colab Notebooks/EjercicioRepasoClase091123.txt")

```

#3.- Genero diccionario con en número de apariciones de cada palabra.
Meto en una lista las frases del archivo.txt, la transformo en otra lista que contenga todas las palabras, elimino las claves duplicadas transformandolas en un set, transformo los elementos de la lista en claves
de un diccionario y finalmente vuelvo a recorrer la lista y sumo 1 cada vez que un valor está repetido.

```python


listapalabras=[]

for line in f:

    listafrases=line.split() #Separa las frases y Me devuelve una lista donde cada elemento es una frase

    for palabra in listafrases:

        listapalabras.append(palabra) #itera las frases y me devuelve una lista donde cada elemento es una palabra.

#quito los respetidos transformando la lista en un set

listapalabrasnorepetidas=set(listapalabras)
#creo un diccionario vacío
diccionariopalabras={}

for i in listapalabrasnorepetidas:
    diccionariopalabras[i]=0 #creo un diccionario de palabras de la lista no repetidas con valor 0 cada una.
    # Y las añado al diccionario
# ahora hay que contar las veces que una palabra aparece dentro de la lista de palabras


for i in listapalabras: # En este bucle cuento las veces que aparece una palabra en las frases,
    for j in listapalabrasnorepetidas: #sumando uno cada vez que aparece una palabra

        if i ==j:
            diccionariopalabras[j]+=1


print(diccionariopalabras)
```

# 4.- Resultado:

```python
print(diccionariopalabras)
print(diccionariopalabras["Chuck"])
```

# 5.- Grafico de barras
Se muestran el número de veces que aparece una palabra en las frases extraídas de la URL.

```python

# Crear un DataFrame de pandas desde el diccionario
df = pd.DataFrame(list(diccionariopalabras.items()), columns=['Clave', 'Valor'])

# Crear un gráfico de barras utilizando pandas y ajustar nombres de las columnas
df.plot(kind='bar', x='Clave', y='Valor', legend=False)
plt.title('Número de apariciones por palabra en el txt de Chuck')
plt.xlabel('Palabra')
plt.ylabel('Nº Apariciones')

plt.show()
```
