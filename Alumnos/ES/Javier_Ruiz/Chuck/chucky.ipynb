```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Abrir el archivo TXT
archivo = open('/Users/User/Documents/GitHub/EDEM_MDA2324/Alumnos/ES/Javier_Ruiz/Chuck/results/chistes.txt', 'r')

# Obtener las palabras del archivo TXT
chistes_text = archivo.read()

# Contar la frecuencia de las palabras
palabras_frecuentes = collections.Counter(chistes_text.split())

# Filtrar solo los sustantivos
sustantivos = []
for palabra in palabras_frecuentes:
    if palabra[0].isalpha() and palabra[-1].isalpha():
        sustantivos.append(palabra)

# Crear un objeto WordCloud
wordcloud = WordCloud(max_font_size=50)

# Generar la nube de palabras
wordcloud.generate(' '.join(sustantivos))

# Crear una figura
fig, ax = plt.subplots()

# Mostrar la nube de palabras
ax.imshow(wordcloud)
ax.axis('off')

# Guardar el grfico
plt.savefig('nube_de_palabras.png')

# Cerrar el archivo TXT
archivo.close()

```
