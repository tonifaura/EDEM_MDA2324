Este conjunto de datos contiene información sobre impagos, factores demográficos, datos crediticios, historial de pagos y extractos de facturas de clientes de tarjetas de crédito en Taiwán desde abril de 2005 hasta septiembre de 2005.

Hay 25 variables:

1. ID: ID de cada cliente
2. LIMIT_BAL: Importe del crédito concedido en dólares NT (incluye crédito individual y familiar/suplementario).
3. SEXO: Sexo (1=masculino, 2=femenino)
4. MATRIMONIO: Estado civil (1=casado, 2=soltero, 3=otros)
5. EDAD: Edad en años
6. PAY_0: Estado de reembolso en septiembre de 2005 (-1=paga debidamente, 1=retraso en el pago durante un mes, 2=retraso en el pago durante dos meses, ... 8=retraso en el pago durante ocho meses, 9=retraso en el pago durante nueve meses y más)
7. PAY_2: Estado de reembolso en agosto de 2005 (escala idéntica a la anterior)
8. PAY_3: situación de reembolso en julio de 2005 (escala idéntica a la anterior)
9. PAY_4: Estado de reembolso en junio de 2005 (escala idéntica a la anterior)
10. PAY_5: Estado de reembolso en mayo de 2005 (escala idéntica a la anterior)
11. PAY_6: Situación del reembolso en abril de 2005 (escala idéntica a la anterior)
12. BILL_AMT1: Importe de la factura en septiembre de 2005 (dólares NT)
13. 13. BILL_AMT2: Importe de la factura en agosto de 2005 (dólares NT)
14. BILL_AMT3: Importe de la factura de julio de 2005 (dólares NT)
15. BILL_AMT4: Importe de la factura de junio de 2005 (dólares NT)
16. BILL_AMT5: Importe de la factura de mayo de 2005 (dólares NT)
17. BILL_AMT6: Importe del extracto de cuenta de abril de 2005 (dólares NT)
18. PAY_AMT1: Importe del pago anterior en septiembre de 2005 (dólares NT)
19. PAY_AMT2: Importe del pago anterior en agosto de 2005 (dólar NT)
20. PAY_AMT3: Importe del pago anterior en julio de 2005 (dólares NT)
21. PAY_AMT4: Importe del pago anterior en junio de 2005 (en dólares neozelandeses)
22. PAY_AMT5: Importe del pago anterior en mayo de 2005 (dólares NT)
23. PAY_AMT6: Importe del pago anterior en abril de 2005 (dólares neozelandeses)
24. default.payment.next.month: Pago por defecto (1=sí, 0=no)
25. EDUCACIÓN: (1=licenciatura, 2=universidad, 3=bachillerato, 4=otros, 5=desconocido, 6=desconocido)

Este conjunto de datos se descargó de https://www.kaggle.com/datasets/uciml/default-of-credit-card-clients-dataset/ y posee la siguiente biografía.

```python
import pandas as pd
import numpy as np
import altair as alt
import plotly.graph_objs as go
```

```python
df = pd.read_csv('/content/datos.csv')
df
```

```python
# Creamos un histograma de las edades de los clientes del banco, y para ello creamos una muestra aleatoria
grafico1 = df.sample(1000)

alt.Chart(grafico1).mark_bar().encode(
  x=alt.X('AGE', bin=True),
  y='count()',
)
```

Para este primer gráfico, se ha seleccionado una muestra aleatoria de 1000 de los clientes que posean tarjetas bancarias. Esto, se ha realizado porque el gráfico seleccionado es un histograma, y por ello la muestra total del dataset es demasiado grande para realizarlo. Como es normal, no hay clientes menores de los 18 años, ya que, no es legal en España.

```python
# Realizamos un gráfico de dispersión para ver como se distribuyen las edades de los clientes del banco, y para ello creamos una muestra aleatoria
g2 = df.sample(1000)
scatter_trace = go.Scatter(
    x=g2['ID'],
    y=g2['AGE'],
    mode='markers',
    marker=dict(color='blue', size=8),
    text=g2['ID']
)

layout = go.Layout(
    title='Gráfico de Dispersión de Edad de Clientes',
    xaxis=dict(title=' ID del Cliente'),
    yaxis=dict(title='Edad')
)

fig = go.Figure(data=[scatter_trace], layout=layout)
fig

```

Para este segundo gráfico, también se ha seleccionado una muestra aleatoria de 1000 de los clientes que posean tarjetas bancarias. Esto, se ha realizado porque el gráfico seleccionado es un gráfico de dispersión, y por ello al poseer aporximadamente 22.000 observaciones, los resultados no son tan vistosos. Realmente, con este gráfico vemos los mismos resultados que en el anterior, pero con un gráfico diferente, esta vez con el propio de la libreria Plotly.

```python
df['MARRIAGE'] = df['MARRIAGE'].astype(str)
def map_marriage_status(value):
    if value == '1':
        return 'casado'
    elif value == '2':
        return 'soltero'
    else:
        return 'otros'

df['MARRIAGE'] = df['MARRIAGE'].apply(map_marriage_status)
```

```python
conteo_valores = df['MARRIAGE'].value_counts()

fig = go.Figure(data=[go.Pie(labels=df['MARRIAGE'].index, values=conteo_valores.values)])
fig.update_layout(
    title='Gráfico de los tipos de estado civil',
)
fig.show()
```

Como se puede observar, se ha realizado un gráfico de tarta para comparar los diferentes niveles de los estados civiles de los clientes. EL grupo dominante con un 53% es el de soltero, mientras que los casados represetan casi la mitad de los datos muestreados con un 45%.

```python
df['EDUCATION'] = df['EDUCATION'].astype(str)
def map_level_edu(value):
    if value == '0':
        return 'NA'
    elif value == '1':
        return 'licenciatura'
    elif value == '2':
        return 'universidad'
    elif value == '3':
        return 'bachillerato'
    elif value == '4':
        return 'otros'
    elif value == '5':
        return 'desconocido'
    elif value == '6':
        return 'desconocido'
    else:
        return 'NA'

df['EDUCATION'] = df['EDUCATION'].apply(map_level_edu)
count_valores = df['EDUCATION'].value_counts()
bar_trace = go.Bar(
    x=count_valores.index,
    y=count_valores.values,
    marker=dict(color='red')
)

layout = go.Layout(
    title='Conteo de Valores en la Variable Entera',
    xaxis=dict(title='Nivel de educacion'),
    yaxis=dict(title='Número de clientes')
)

fig = go.Figure(data=[bar_trace], layout=layout)
fig.show()
```

En este caso, se ha realizado un grafico de barras en el que se muestran los diferentes niveles de educación que poseen cada uno de los clientes. Para la realización de este, se han unido los clientes que poseeían un valor de 5 y 6, ya que, ambos eran desconocidos. Además se ha renombrado a los diferentes grupos para así hacer mucho mas visual el gráfico.
