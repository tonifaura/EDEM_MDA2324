```python
# This script reads  chuck.txt file and count the times a word repeats:
# Open the file in read mode
text = open("chuck/results/results.txt", "r")

# Create an empty dictionary
chucktionary = dict()

# Loop through each line of the file
for line in text:
	# Remove the leading spaces and newline character
	line = line.strip()

	# Convert the characters in line to
	# lowercase to avoid case mismatch
	line = line.lower()

	# Split the line into words
	words = line.split(" ")

 		# Iterate over each word in line
	for word in words:
		# Remove all non alpha-numeric characters from the word
		import string # string library needed for passing functions to process
		# strings
		my_string = word
		# We create a translation table:
		mapping_table = str.maketrans('', '', string.punctuation)
		# And we re-assign the 'cleaned' string to word again by passing
		# mapping_table as an argument to .translate() method:
		word = my_string.translate(mapping_table)
		# Check if the word is already in dictionary
		if word in chucktionary:
			# Increment count of word by 1
			chucktionary[word] = chucktionary[word] + 1
		else:
			# Add the word to dictionary with count 1
			chucktionary[word] = 1

# Print the contents of dictionary
for key in list(chucktionary.keys()):
	print(key, ":", chucktionary[key])
```

```python
# Too many words. Let's create a dictionary sorted by values and pick the top-5:
# First, let's sort it by values. This will create a list that we'll convert
# back to a dictionary.
sorted_list = sorted(chucktionary.items(), key = lambda x:x[1], reverse = True)
chucktionary_sorted = dict(sorted_list)
print(chucktionary_sorted)
```

```python
# Let's plot how many times each word repeats in the text file
import matplotlib.pyplot as plt

plt.bar(range(10), list(chucktionary_sorted.values())[0:10], align='center')
plt.xticks(range(10), list(chucktionary_sorted.keys())[0:10])
plt.xlabel("Words")
plt.ylabel("No. of times a word repeats")
plt.title("TOP-10 WORDS")

plt.show()
```
