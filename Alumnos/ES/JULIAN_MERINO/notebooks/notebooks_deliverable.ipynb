```python
import pandas as pd
df = pd.read_csv("/content/raw-data-kaggle.csv",sep = ';')
df
```

```python
# dataset loaded previously from raw-data-kaggle.csv

# create a smaller dataset

df_2 = df.sample(1000)

# plot the dataset, referencing dataframe column names
import altair as alt
alt.Chart(df_2).mark_point().encode(
  x='distance (m)',
  y='elapsed time (s)',
).interactive()
```
