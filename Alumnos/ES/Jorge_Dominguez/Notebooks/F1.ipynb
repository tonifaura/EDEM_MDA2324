```python
numerical_columns = f1.select_dtypes(include=['float64', 'int64']).columns


plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(f1[numerical_columns].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Heatmap for F1 Drivers Dataset')
plt.show()
```

Mapa de correlaciones para ver qué relación guardan las variables de forma general. En este caso, el número de campeonatos está altamente correlacionado con Pole_Positions (0.87), Race_Wins (0.92), y Podiums (0.81), cosa obvia la verdad, ya que si sales en pole es más seguro que ganes, y si ganas mucho es más probable que tengas un campeonato, al igual que si haces muchos podios.

Race_Entries, Race_Starts, Pole_Positions, Race_Wins, Podiums, y Fastest_Laps tienen fuertes correlaciones positivas entre sí, lo que sugiere que los pilotos que participan en más carreras tienden a tener más salidas, poles, victorias, podios, y vueltas rápidas. Es un efecto causal simple

Los puntos están fuertemente correlacionados con las Carreras_Ganadas (0,93) y los Podios (1,00), lo que es de esperar ya que ganar carreras y terminar en el podio hace que los pilotos ganen más puntos.

No es el análisis más exhaustivo pero no hace falta más.

```python
plt.figure(figsize=(14, 7))
sns.histplot(f1['Race_Entries'], bins=50, kde=True)
plt.title('Distribution of Race Entries')
plt.xlabel('Number of Race Entries')
plt.ylabel('Frequency')
plt.axvline(f1['Race_Entries'].mean(), color='r', linestyle='--', label=f"Mean Entries: {f1['Race_Entries'].mean():.2f}")
plt.legend()
plt.show()
```

Este histograma muestra la distribución de frecuencias del número de participaciones por piloto.
La línea discontinua roja indica la media de participaciones, lo que proporciona una referencia del número de carreras en las que participa un piloto medio.
La mayoría de los pilotos participan en menos de 50 carreras, con un descenso significativo a medida que aumenta el número de participaciones. El Nano como es especial cuenta con 359 carreras en el dataframe

```python
plt.figure(figsize=(14, 7))
sns.countplot(x='Championships', data=f1[f1['Championships'] > 0])
plt.title('Number of Championships Won by Drivers')
plt.xlabel('Number of Championships')
plt.ylabel('Number of Drivers')
plt.yscale('log')
plt.show()
```

Este gráfico muestra cuántos campeonatos han ganado los pilotos.
El eje Y está en una escala logarítmica para mostrar mejor la distribución, ya que muy pocos pilotos han ganado varios campeonatos.
Es evidente que los campeonatos son un logro poco frecuente, ya que la mayoría de los pilotos no han ganado ninguno. El Nano tiene 2, si hubiera fichado por Red Bull en 2011 tendría más.

```python
plt.figure(figsize=(14, 7))
nationality_counts = f1['Nationality'].value_counts().head(20)
sns.barplot(x=nationality_counts.index, y=nationality_counts.values)
plt.title('Top 20 Nationalities Represented in F1')
plt.xlabel('Nationality')
plt.ylabel('Number of Drivers')
plt.xticks(rotation=90)
plt.show()
```

El diagrama de barras muestra las 20 nacionalidades más representadas por el número de conductores en el conjunto de datos.
El Reino Unido tiene la mayor representación, seguido de otros países.
Esto da una idea de qué países han sido históricamente más activos o han tenido más éxito en la Fórmula 1. Ojalá más pilotos Españoles, pero es un deporte caro y a los ingleses les va más lo de hacer karts.
La muestra de los pilotos de Estados Unidos es un poco falsa, pues hubo una epoca que la IndyCar (la F1 de USA) contaba como una división de la F1, por lo que el número está claramente inflado. Y aún estando inflado no hay ningún piloto bueno en ese país.

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(x='Points_Per_Entry', y='Race_Wins', hue='Active', data=f1)
plt.title('Points Per Entry vs. Race Wins')
plt.xlabel('Points Per Entry')
plt.ylabel('Race Wins')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(f1[f1['Win_Rate'] > 0]['Win_Rate'], bins=50, kde=True)
plt.title('Distribution of Win Rates for Drivers')
plt.xlabel('Win Rate')
plt.ylabel('Frequency')
plt.axvline(f1['Win_Rate'].mean(), color='r', linestyle='--', label=f"Mean Win Rate: {f1['Win_Rate'].mean():.4f}")
plt.legend()
plt.show()
```

El histograma muestra la distribución de los porcentajes de victorias entre los pilotos que han ganado al menos una carrera.
El porcentaje medio de victorias está marcado con una línea discontinua roja, que muestra el porcentaje medio de victorias.

```python
plt.figure(figsize=(14, 7))
sns.histplot(f1['Years_Active'], bins=30, kde=True)
plt.title('Distribution of Years Active Among Drivers')
plt.xlabel('Years Active')
plt.ylabel('Frequency')
plt.axvline(f1['Years_Active'].mean(), color='r', linestyle='--', label=f"Mean Years Active: {f1['Years_Active'].mean():.2f}")
plt.legend()
plt.show()
```

Este gráfico muestra cuántos años permanecen activos los pilotos en la Fórmula 1.
Se indica el número medio de años en activo, lo que da una idea de la duración media de la carrera.
La distribución está sesgada a la derecha, lo que sugiere que la mayoría de los pilotos tienen una carrera relativamente corta en la F1, con menos pilotos que mantienen carreras más largas. Esto pasa porque conducir como Fernando Alonso es dificil, pues es de los que más años tiene en activo con 19 años.

```python
plt.figure(figsize=(14, 7))
sns.boxplot(x='Decade', y='Points_Per_Entry', data=f1)
plt.title('Points Per Entry by Decade')
plt.xlabel('Decade')
plt.ylabel('Points Per Entry')
plt.show()
```

El boxplot compara los puntos obtenidos por participante en las distintas décadas.
Ofrece una perspectiva histórica de cómo ha cambiado la puntuación y quizá la competitividad de este deporte con el paso del tiempo.
Hay variabilidad en los puntos por entrada dentro de cada década, y parece que la mediana de puntos por entrada ha variado significativamente con el tiempo. Esto es porque ahora se regalan más puntos por carrera (puntúan los 10 primeros) con más puntos por posición, mientras que antaño puntuaban los 6 primeros con menos puntos por posición. A su vez, la aparición de puntos por vueltas rápidas y Sprint Races aunmenta los puntos totales en los últimos años.

```python
top_champions = f1.sort_values(by='Championships', ascending=False).head(10)

plt.figure(figsize=(14, 7))
sns.barplot(x='Driver', y='Championships', data=top_champions)
plt.title('Top 10 Drivers by Championships Won')
plt.xlabel('Driver')
plt.ylabel('Total Championships')
plt.xticks(rotation=90)
plt.show()
```

Este gráfico de barras compara el número total de campeonatos ganados por los 10 mejores pilotos.
Fernando Alonso estaría en la lista si en 2012 Vettel hubiera corrido con un coche legal en Brasil, y si PedroDLR se hubiese estampado contra él en su HRT(coche español que no corría)

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Driver', y='Race_Wins', data=top_champions)
plt.title('Top 10 Drivers by Race Wins')
plt.xlabel('Driver')
plt.ylabel('Total Race Wins')
plt.xticks(rotation=90)
plt.show()
```

El gráfico ilustra el número total de victorias de los mismos 10 pilotos.
Como podemos ver, con el paso del tiempo, hace falta ganar más carreras para proclamarse campeón, esto debido a que a día de hoy hay más circuitos por temporada.

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Driver', y='Podiums', data=top_champions)
plt.title('Top 10 Drivers by Podium Finishes')
plt.xlabel('Driver')
plt.ylabel('Total Podiums')
plt.xticks(rotation=90)
plt.show()
```

Esta gráfico compara el total de podios de estos pilotos. Lo mismo que antes

```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Driver', y='Points', data=top_champions)
plt.title('Top 10 Drivers by Points Scored')
plt.xlabel('Driver')
plt.ylabel('Total Points')
plt.xticks(rotation=90)
plt.show()
```

El gráfico de barras muestra el total de puntos obtenidos por cada uno de los 10 primeros pilotos. Como hemos mencionado antes, actualmente es más fácil obtener puntos y por lo tanto necesitas más para ganar el campeonato.
