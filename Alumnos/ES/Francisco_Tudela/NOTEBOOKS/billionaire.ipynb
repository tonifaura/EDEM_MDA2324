```python
bill2 = billns.groupby('personName')['finalWorth'].sum().sort_values(ascending=False)

richest_p = bill2.head(10)

plt.figure(figsize=(12, 8))
richest_p.plot(kind='bar', color='gold')
plt.title('Top 10 $$$')
plt.xlabel('Who?')
plt.ylabel('How much??')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
gender_counts = billns['gender'].value_counts()

fig3 = px.pie(gender_counts, labels=gender_counts.index, values=gender_counts.values,
              title="Gender Distribution of Billionaires",
              color_discrete_sequence=px.colors.qualitative.Plotly)
fig3.update_traces(marker=dict(line=dict(color='white', width=2)))
fig3.update_layout(showlegend=False)
fig3.update_traces(hole=0.4)

gender_labels = gender_counts.index
fig3.add_annotation(
    text="<b>Gender</b>",
    x=0.5,
    y=0.5,
    showarrow=False,
    font=dict(size=15),
)
fig3.add_annotation(
    text=gender_labels[0],
    x=0.53,
    y=0.1,
    showarrow=False,
    font=dict(size=12),
)
fig3.add_annotation(
    text=gender_labels[1],
    x=0.47,
    y=0.81,
    showarrow=False,
    font=dict(size=12),
)

fig3.show()
```
