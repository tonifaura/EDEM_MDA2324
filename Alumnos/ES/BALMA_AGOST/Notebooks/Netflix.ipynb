```python
plt.figure(figsize=(14, 8)) # Tamaño figura

# Obtener las frecuencias de cada país y sus etiquetas
country_counts = df['Country'].value_counts()
countries = country_counts.index

# Colores en gama de azules
colors = sns.color_palette('Blues', len(countries))

# Crear el gráfico de barras con colores diferentes para cada barra
plt.bar(countries, country_counts, color=colors)

plt.title('Distribution of Netflix Users by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()




```

 Según este conjunto de datos

- Estados Unidos tiene el mayor número de usuarios de Netflix.
- España ocupa el segundo lugar en número de usuarios de Netflix.
- Canadá ocupa el tercer lugar en número de usuarios de Netflix.


```python


# Calcular las frecuencias de cada tipo de suscripción
sub_counts = df['Subscription Type'].value_counts()

plt.figure(figsize=(8, 8)) # Tamaño figura

# Colores en tonos azules
colors = ['#1f77b4', '#aec7e8', '#98c2c7', '#7b97c5', '#6b66ff']

# Crear el gráfico de torta con un agujero en el centro
plt.pie(sub_counts, labels=sub_counts.index, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops=dict(width=0.6))

plt.title('Distribution of Subscription Types')
plt.show()
```

La mayoría de los usuarios de Netflix, un 40%, ha optado por el plan de suscripción Básico.
Alrededor del 30,7% de los usuarios ha elegido el plan Estándar.
El 29,3% restante se ha suscrito al plan Premium.
Los planes más baratos suelen atraer a más clientes, ya que la mayoría de los usuarios de Netflix se han suscrito al plan Básico, que es la opción más barata.

```python
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Netflix Userbase.csv')

# Calcular la cuenta de suscripciones por país y tipo de suscripción
subscription_counts = data.groupby(['Country', 'Subscription Type'])['User ID'].count().unstack()

# Colores para cada tipo de suscripción
colors = ['#1f77b4', '#aec7e8', '#98c2c7']

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras apiladas
subscription_counts.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6))

# Configurar el título del gráfico
plt.title('Subscription Plan Counts by Country')

plt.xlabel('Country')
plt.ylabel('Number of Subscriptions')
plt.legend(title='Subscription Type')
plt.xticks(rotation=45) # Poner nombre de los paises en horizontal
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Calcular las frecuencias de cada tipo de dispositivo
device_counts = df['Device'].value_counts()

plt.figure(figsize=(8, 8)) # Tamaño figura

# Colores en tonos azules
colors = ['#1f77b4', '#aec7e8', '#98c2c7', '#7b97c5', '#6b66ff']

# Crear el gráfico de torta con un agujero en el centro
plt.pie(device_counts, labels=device_counts.index, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops=dict(width=0.6))

plt.title('Device Distribution')
plt.show()
```

El uso de diferentes dispositivos para acceder a Netflix está distribuido de forma relativamente uniforme. Alrededor de una cuarta parte de los usuarios acceden a Netflix desde un ordenador portátil, una tableta, un smartphone o un Smart TV, respectivamente. Esto sugiere que Netflix está diseñado para ser fácilmente accesible a través de diferentes dispositivos y plataformas.

```python
subscription_revenue = df.groupby('Subscription Type')['Monthly Revenue'].sum().reset_index()

# Definir gama de azules
colors = ['#1f77b4', '#aec7e8', '#98c2c7', '#7b97c5', '#6b66ff']

# Crear un gráfico de barras azules
plt.figure(figsize=(10, 6))
bars = plt.bar(subscription_revenue['Subscription Type'], subscription_revenue['Monthly Revenue'], color=colors)

plt.xlabel('Subscription Type')
plt.ylabel('Total Monthly Revenue')
plt.title('Total Monthly Revenue by Subscription Type')
plt.xticks(rotation=45) # Para que se lea bien


# Añadir etiquetas en las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')


plt.show()
```

Este gráfico muestra la distribución de los tipos de suscripción, lo que puede ser útil para determinar qué planes de suscripción son más populares entre los usuarios. Podemos observar, que claramente el plan más demandado es el Basico, seguidamente el Estándar y por último, el Premium.
