```python
import numpy as np
import nltk
import matplotlib.pyplot as plt
nltk.download('stopwords')
```

```python
text_document = open('chucky.txt', 'r')
string = text_document.read()
```

```python
string_sep = string.split()
```

```python
stopwords = nltk.corpus.stopwords.words("english")
```

```python
nouns = []
```

```python
for i in range(len(string_sep)):
    if string_sep[i-1] not in stopwords and string_sep[i-1].isalpha():
        nouns.append(str(string_sep[i-1]).lower())       
```

```python
plt.hist(nouns, orientation='horizontal')
plt.show()
```
