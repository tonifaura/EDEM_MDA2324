```python
df = pd.read_csv('billionaires.csv')

# Calcular la media y la desviación estándar de la edad
mu = df['age'].mean()
std = df['age'].std()

# Crear un rango de valores para la distribución normal
x = np.linspace(mu - 3*std, mu + 3*std, 100)

# Calcular la función de densidad de probabilidad de la distribución normal
pdf = norm.pdf(x, mu, std)

# Calcular los porcentajes de datos dentro de ciertos intervalos
prob_under_minus1 = norm.cdf(mu - std, mu, std)
prob_over_1 = 1 - norm.cdf(mu + std, mu, std)
between_prob = 1 - (prob_under_minus1 + prob_over_1)

# Gráfico de densidad de edad de personas más ricas
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, color='skyblue', label='Densidad de Edad')

# Áreas sombreadas para representar los porcentajes
plt.fill_between(x, pdf, where=(x < mu - std), color='red', alpha=0.3, label=f'{round(prob_under_minus1*100, 2)}% de datos < -1')
plt.fill_between(x, pdf, where=(x > mu + std), color='red', alpha=0.3, label=f'{round(prob_over_1*100, 2)}% de datos > 1')
plt.fill_between(x, pdf, where=((x >= mu - std) & (x <= mu + std)), color='blue', alpha=0.3, label=f'{round(between_prob*100, 2)}% de datos entre -1 y 1')

# Etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.title('Gráfico de Densidad de Edad de Personas más Ricas')

# Mostrar el gráfico
plt.show()
```

El gráfico de densidad nos muestra que la mayoría de las personas más ricas se concentran alrededor de la media o mediana de la distribución. A su vez, conforme te vas acercando o alejando de la media, la densidad de las personas disminuye paulatinamente. A priori, las personas más jóvenes y más mayores no se encuentran entre los más ricos, sino que queda en los grupos de edad intermedia.

### 3.3.3 Prueba de Levene
A continuación se realizará la homogeneidad de varianzas entre distintos grupos. En primer lugar será escogido el finalWorth de los billonarios, así como las categorías industriales en las que se encuentran incluidos.

```python


# Separar los datos en grupos basados en una variable categórica (por ejemplo, 'category')
grupos = [df[df['category'] == categoria]['finalWorth'] for categoria in df['category'].unique()]

# Realizar la prueba de Levene
estadistico, valor_p = levene(*grupos)

# Imprimir los resultados
print("Estadístico de Levene:", estadistico)
print("Valor p:", valor_p)

# Interpretar los resultados
alpha = 0.05
if valor_p > alpha:
    print("No hay evidencia suficiente para rechazar la hipótesis nula. Las varianzas son homogéneas.")
else:
    print("Se rechaza la hipótesis nula. Las varianzas no son homogéneas.")
```

### 3.3.4 Contraste de Hipótesis

A continuación se realiza el contraste de hipótesis con t de Student para poder comparar dos medias poblacionales. En este caso, se va a comparar dos grupos independientes que no se encuentran relacionados entre sí como son la riqueza final entre hombres y mujeres. De esta forma se quiere conocer si existe diferencia significativa en la riqueza por género.

```python
# Prueba t de Student para corroborar si existen diferencias significativas entre hombres y mujeres.

# CONTRASTE DE HIPÓTESIS:

# H0: Hipótesis Nula. Hay una diferencia significativa en la riqueza final entre hombres y mujeres.
# H1: Hipótesis Alternativa. No hay suficiente evidencia para concluir que hay una diferencia significativa en la riqueza final entre hombres y mujeres.

riqueza_hombres = df[df['gender'] == 'M']['finalWorth']
riqueza_mujeres = df[df['gender'] == 'F']['finalWorth']

# Realizar la prueba t de Student para muestras independientes
statistic, p_value = stats.ttest_ind(riqueza_hombres, riqueza_mujeres)

print(f"Estadístico de prueba t: {statistic}")
print(f"Valor p:{p_value}")

# Resultados del análisis:
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula. Hay una diferencia significativa en la riqueza final entre hombres y mujeres.")
else:
    print("No podemos rechazar la hipótesis nula. No hay suficiente evidencia para concluir que hay una diferencia significativa en la riqueza final entre hombres y mujeres.")
```

Como bien indica el resultado de la prueba, el p value es de 0.8878335242236445, por lo que se acepta la hipótesis nula ("Hay una diferencia significativa en la riqueza final entre hombres y mujeres.). No existe suficiente evidencia estadística para afirmar que no existe una diferencia en la riqueza final entre hombres y mujeres.

### 3.3.5 Análisis ANOVA y Tukey

En este apartado se presenta un **análisis de varianza (ANOVA)** para evaluar las diferencias entre las medias de dos o más grupos en una variable dependiente. 
El primer ANOVA se realiza para examinar la **relación entre la edad y la category (industria)**. Específicamente, se busca determinar si existe una diferencia estadísticamente significativa en la edad promedio entre las diferentes categorías industriales.

```python
df = pd.read_csv('billionaires.csv')

# Eliminar filas con valores NaN en la columna de edad
df = df.dropna(subset=['age'])

# Modelo ANOVA
modelo_anova = ols('age ~ category', data=df).fit()

# Realizar ANOVA
tabla_anova = sm.stats.anova_lm(modelo_anova, typ=2)

# Realizar un análisis de Tukey
comparaciones_tukey = sm.stats.multicomp.pairwise_tukeyhsd(df['age'], df['category'])

print(tabla_anova)
print(comparaciones_tukey)

```

**Conclusión del análisis ANOVA y prueba de Tukey**

Tras realizar el primer estudio, el análisis de varianza (ANOVA) revela una diferencia significativa en la variable de edad entre al menos dos de las industrias estudiadas. Se ha hecho un análisis post-hoc, a través de **Tukey**.  El resultado concluye que "Technology" tiene una edad promedio significativamente más baja que la industria de "Automotive" y también que la industria "Sports". Esto significa que, en general, las personas que trabajan en tecnología tienden a ser más jóvenes que las personas que trabajan en la industria automotriz o en deportes.

**El segundo ANOVA** 

Se realiza para analizar la relación entre el valor final (finalWorth) y las industrias estudiadas. Se busca determinar si existe una diferencia estadísticamente significativa en el valor final promedio entre las diferentes categorías industriales.

```python
# Ajustar un modelo ANOVA
modelo_anova = ols('finalWorth ~ category', data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_anova, typ=2)

# Realizar un análisis de Tukey
comparaciones_tukey = sm.stats.multicomp.pairwise_tukeyhsd(df['finalWorth'], df['category'])

# Imprimir los resultados del ANOVA
print("Resultados del ANOVA:")
print(tabla_anova)

# Imprimir los resultados del análisis de Tukey
print("\nComparaciones de Tukey:")
print(comparaciones_tukey)
```

**Conclusión del análisis ANOVA y prueba de Tukey**

El análisis ANOVA y la prueba post-hoc de Tukey indican que existen diferencias estadísticamente significativas entre las medias de las categorías industriales en la variable "finalWorth".
