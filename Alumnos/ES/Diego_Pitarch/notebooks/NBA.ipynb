###An√°lisis de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/all_seasons.csv")
df
```

```python
df.dtypes
```

##An√°lisis de datos NBA

```python

# Sacar cuantas veces hay de cada altura
counts = df['player_height'].value_counts()

counts = counts.sort_index()

#Creo gr√°fico
plt.bar(counts.index, counts.values)
plt.title('Distribuci√≥n de Alturas de Jugadores')
plt.xlabel('Altura del Jugador')
plt.ylabel('Cantidad de Jugadores')

plt.xticks(rotation=90)
plt.show()
```

###En este primer gr√°fico podemos ver la cantidad de jugadores que hay dependiendo de su altura. Observamos que por debajo de 180cm y por encima de 220cm, apenas encontramos jugadores. Estos se concentran enmedio de las dos medidas anteriores, llegando a la cantidad m√°xima en los 206cm.

```python
from matplotlib import pyplot as plt
df['pts'].plot(kind='hist', bins=20, title='points')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

###Como podemos ver en el gr√°fico, la gran mayor√≠a de jugadores anotan muy pocos puntos de media por partid, podemos decir que solo alrededor de 250 jugadores tienen una media mayor a 20 puntos

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='draft_round', y='pts', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

###La forma m√°s habitual de entrar dentro de la NBA es a partir del Draft, los jugadores se presentan como seleccionables y los equipos van eligiendo conforme a sus necesidades. Normalmente en las primeras posiciones suelen ser elegidos los mejores jugadores. En este gr√°fico vemos que cuanto mas arriba son elegidos conlleva que anotan mas puntos.

```python
# Ordenar gr√°ficos de mayor a menor anotador
top_10_pts = df.sort_values(by='pts', ascending=False).head(17)

#Creo gr√°fico
plt.bar(top_10_pts['player_name'], top_10_pts['pts'])

plt.title('Los 10 M√°ximos anotadores')
plt.xlabel('Jugador')
plt.ylabel('puntos')
plt.xticks(rotation=90)

plt.show()
```

### En este gr√°fico podemos observar los  10 m√°ximos anotadores medios de la NBA

```python
#  Ordenar gr√°ficos de mayor a menor reboteador
top_10_reb = df.sort_values(by='reb', ascending=False).head(17)

# Creo gr√°fico de barras
plt.bar(top_10_reb['player_name'], top_10_reb['reb'])

plt.title('Los 10 M√°ximos Reboteadores')
plt.xlabel('Jugador')
plt.ylabel('Rebotes')
plt.xticks(rotation=90)

plt.show()

```

### En este gr√°fico podemos observar los  10 m√°ximos reboteadores medios de la NBA

```python

# Ordenar gr√°ficos de mayor a menor asistente
top_10_ast = df.sort_values(by='ast', ascending=False).head(25)

# Creo gr√°fico de barras
plt.bar(top_10_ast['player_name'], top_10_ast['ast'])
plt.title('Los 10 M√°ximos asistentes')
plt.xlabel('Jugador')
plt.ylabel('asistencias')

plt.xticks(rotation=90)
plt.show()
```

### En este gr√°fico podemos observar los  10 m√°ximos asistentes medios de la NBA

```python


# Creo nueva variable 'country_category' que indique si es "USA" o "Resto"
df['country_category'] = df['country'].apply(lambda x: 'USA' if x == 'USA' else 'Resto')
country_counts = df['country_category'].value_counts()

# Crear un gr√°fico circular
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, counterclock=False)

plt.title('Distribuci√≥n de Pa√≠ses (USA vs. Resto del mundo)')
plt.axis('equal')
plt.show()

```

###En este gr√°fico circular podemos observar la distribuci√≥n de las nacionalidades de los jugadores NBA, de estos el 83.5% son am√©ricanos y el resto de nacionales cuenta con el 16.5%.
