### *Pie chart métodos de ejecución*





```python
import pandas as pd
import matplotlib.pyplot as plt

# Especifica la ruta de tu archivo CSV
ruta_archivo_csv = '/content/database.csv'

# Carga el archivo CSV en un DataFrame de pandas
dataframe = pd.read_csv(ruta_archivo_csv)

# Cuenta los valores únicos en la columna 'Method'
conteo_metodos = dataframe['Method'].value_counts()

# Calcula los porcentajes de cada categoría
total = sum(conteo_metodos)
porcentajes = [(count / total) * 100 for count in conteo_metodos]

# Define los colores que deseas usar
colores = ['skyblue', 'lightgreen', 'orange', 'red', 'pink']

# Crea un pie chart con los colores definidos y sin etiquetas
plt.figure(figsize=(10, 8))  # Tamaño del gráfico
pie = plt.pie(conteo_metodos, startangle=140, colors=colores)

# Ajusta manualmente las posiciones de las etiquetas para evitar superposiciones
plt.gca().set_aspect('equal')  # Aspecto circular
plt.subplots_adjust(left=0.1, right=0.65, top=0.8, bottom=0.1)  # Ajuste de posición

# Agrega una leyenda con los colores, etiquetas y porcentajes correspondientes
leyenda = [f"{label} ({porcentaje:.1f} %)" for label, porcentaje in zip(conteo_metodos.index, porcentajes)]
plt.legend(leyenda, title="Método", loc="center left", bbox_to_anchor=(1, 0.5))

# Muestra el gráfico
plt.show()

```

```python
# Crear tabla de frecuencias con el total
tabla_frecuencias = pd.crosstab(index=dataframe['Method'], columns='count', margins=True)

# Mostrar la tabla de frecuencias con el total en formato tabular
from IPython.display import display
display(tabla_frecuencias)

```

Basándonos en los resultados del pie chart, podemos extraer las siguientes conclusiones:

1. **Dominancia de la inyección letal:** El método de ejecución más común en el conjunto de datos es la inyección letal, representando aproximadamente el 87.9% de todas las ejecuciones registradas. Esta alta proporción indica que la inyección letal es el método preferido o más utilizado en comparación con los demás.

2. **Baja incidencia de otros métodos:** Los otros métodos de ejecución, tienen una presencia significativamente menor en el conjunto de datos. Estos métodos combinados representan solo el 12.1% restante de las ejecuciones, con porcentajes muy bajos para cada uno.

3. **Métodos alternativos poco frecuentes:** La cámara de gas, el pelotón de fusilamiento y la horca tienen una representación mínima en el conjunto de datos, cada uno con menos del 1% de las ejecuciones. Esto sugiere que estos métodos son menos comunes o menos utilizados en comparación con la inyección letal.

En resumen, el pie chart muestra una clara dominancia de la inyección letal como el método preferido de ejecución en el conjunto de datos, con una baja incidencia de otros métodos alternativos.

MEDIDAS DE DISPERSIÓN:

```python
import numpy as np

# Extraemos la columna 'Age' del DataFrame
edades = dataframe['Age']

# i) Rango
rango = np.ptp(edades)

# ii) Rango Intercuartílico (IQR)
Q3, Q1 = np.percentile(edades, [75 ,25])
iqr = Q3 - Q1

# iii) Varianza
varianza = np.var(edades)

# iv) Desviación típica
desviacion_tipica = np.std(edades)

# v) Coeficiente de variación
coeficiente_variacion = (desviacion_tipica / np.mean(edades)) * 100

# Mostramos los resultados
print("Rango:", rango)
print("Rango Intercuartílico (IQR):", iqr)
print("Varianza:", varianza)
print("Desviación Típica:", desviacion_tipica)
print("Coeficiente de Variación:", coeficiente_variacion)
```

Estas son las conclusiones que podemos sacar de las medidas estadísticas calculadas para la variable 'Age':

1. **Rango:** El rango de edades es de 55 años, lo que significa que la diferencia entre la edad máxima y la mínima en el conjunto de datos es de 55 años. Esta medida nos da una idea de la amplitud de las edades en el conjunto de datos.

2. **Rango Intercuartílico (IQR):** El rango intercuartílico es de 13.0 años. Esto indica que el 50% de las edades se encuentran dentro de un rango de 13 años, desde el primer cuartil (Q1) hasta el tercer cuartil (Q3). Es útil para medir la dispersión de las edades excluyendo los valores extremos.

3. **Varianza:** La varianza de las edades es de aproximadamente 89.65. La varianza es una medida de dispersión que indica cuán dispersos están los datos con respecto a la media. Valores más altos de varianza indican una mayor dispersión de los datos.

4. **Desviación Típica:** La desviación típica de las edades es de aproximadamente 9.47 años. La desviación estándar es otra medida de dispersión que indica la cantidad de variación o dispersión de un conjunto de datos. Es la raíz cuadrada de la varianza y se expresa en las mismas unidades que los datos originales.

5. **Coeficiente de Variación:** El coeficiente de variación es del 22.83%. Esta medida relativa de dispersión indica la variabilidad de las edades en relación con su media. Un coeficiente de variación más alto indica una mayor variabilidad relativa, mientras que un valor más bajo indica una menor variabilidad relativa.

En resumen, estas medidas nos dan una idea de la dispersión y la variabilidad de las edades en el conjunto de datos. Las edades parecen estar relativamente dispersas alrededor de la media, con un coeficiente de variación del 22.83%, lo que sugiere una moderada variabilidad relativa en comparación con la media.

Los puntos individuales fuera de los bigotes se consideran valores atípicos o extremos. Estos son puntos que caen más allá de 1.5 veces el rango intercuartílico por encima del tercer cuartil o por debajo del primer cuartil. ### *Contraste hipótesis sobre la normalidad de la variable 'Age'*

```python
import pandas as pd
from scipy.stats import shapiro

# Especifica la ruta de tu archivo CSV
ruta_archivo_csv = '/content/database.csv'

# Carga el archivo CSV en un DataFrame de pandas
dataframe = pd.read_csv(ruta_archivo_csv)

# Selecciona la columna de interés para el contraste de hipótesis
columna_de_interes = dataframe['Victim Count']

# Realiza la prueba de Shapiro-Wilk
stat, p_valor = shapiro(columna_de_interes)

# Define el nivel de significancia (alpha)
alpha = 0.05

# Compara el p-valor con alpha para tomar la decisión
if p_valor > alpha:
    print("No se rechaza la hipótesis nula: la muestra parece seguir una distribución normal (p-valor=%.4f)" % p_valor)
else:
    print("Se rechaza la hipótesis nula: la muestra no sigue una distribución normal (p-valor=%.4f)" % p_valor)
```

El test de Shapiro-Wilk es una prueba de normalidad donde la hipótesis nula afirma que una muestra proviene de una población normalmente distribuida. # ANÁLISIS DE LA VARIANZA (ANOVA)

Para realizar un análisis ANOVA adecuado debemos tener en cuenta los siguientes puntos:

- Datos independientes.
Consideramos que los datos que tenemos tienen independencia entre ellos, ya que se produzca una ejecucción, no tiene nada que ver con ninguna otra ejeccución de nuestra base de datos.

- Normalidad: Como hemos podido ver antes en el gráfico de densidad, los datos siguen una distribución normal.

- Homogeneidad de varianzas: Como hemos podido observar en el boxplot de los métodos de ejecución, no tenemos homogeneidad en las varianzas, ya que los métodos de inyección letal y electrocución son mucho más utilizados que el resto. Sin embargo, hemos considerado realizar igualmente la prueba ANOVA, ya que si eliminamos estos dos métodos, los tres restantes tienen muy pocas muestras y el análisis sería casi irrelevante.

- Niveles de medición: Estamos utilizando un método nominal ya que nuestras categorias no son excluyentes. [MRDB]:?

- Independencia de errores: Asumimos que los errores de las observaciones son independientes entre sí, los errores de una observación no deben influir en otras observaciones.

- Tamaño de muestra adecuado: Consideramos nuestro tamaño muestral adecuado, ya que tenemos 1442 observaciones.

- Diseño experimental adecuado: Consideramos nuestro diseño experimental adecuados ya que estamos haciendo el experimento con 5 métodos de ejecución comparando con todas las edades de las personas que hay en la muestra.

```python
import pandas as pd
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

# Especifica la ruta de tu archivo CSV
ruta_archivo_csv = '/content/database.csv'

# Carga el archivo CSV en un DataFrame de pandas
dataframe = pd.read_csv(ruta_archivo_csv)

# Filtra el DataFrame para incluir solo los tipos de muerte especificados
tipos_de_muerte = ['Firing Squad', 'Electrocution', 'Gas Chamber', 'Lethal Injection', 'Hanging']
dataframe_filtrado = dataframe[dataframe['Method'].isin(tipos_de_muerte)]

# Realiza un ANOVA
resultado_anova = f_oneway(
    dataframe_filtrado['Age'][dataframe_filtrado['Method'] == 'Firing Squad'],
    dataframe_filtrado['Age'][dataframe_filtrado['Method'] == 'Electrocution'],
    dataframe_filtrado['Age'][dataframe_filtrado['Method'] == 'Gas Chamber'],
    dataframe_filtrado['Age'][dataframe_filtrado['Method'] == 'Lethal Injection'],
    dataframe_filtrado['Age'][dataframe_filtrado['Method'] == 'Hanging']
)

# Muestra el resultado del ANOVA
print("Estadística F:", resultado_anova.statistic)
print("Valor p:", resultado_anova.pvalue)

# Verifica si hay diferencias significativas
if resultado_anova.pvalue < 0.05:
    print("Hay diferencias significativas entre los grupos.")
else:
    print("No hay diferencias significativas entre los grupos.")

# Realiza un análisis de Tukey
tukey_results = pairwise_tukeyhsd(dataframe_filtrado['Age'], dataframe_filtrado['Method'])

# Muestra el resumen del análisis de Tukey
print("\nResumen del análisis de Tukey:")
print(tukey_results)

# Realiza el gráfico simultáneo de Tukey
tukey_results.plot_simultaneous()

# Obtiene el gráfico actual
ax = plt.gca()

# Itera sobre las barras del gráfico y cambia su color
for i, line in enumerate(ax.lines):
    if i % 2 == 0:  # Las barras pares son las barras de error
        line.set_color('red')

# Ajusta el título y etiquetas de los ejes
plt.title('Comparación de Edades entre Métodos de Ejecución')
plt.xlabel('Edad')
plt.ylabel('Método de Ejecución')

# Muestra el gráfico
plt.show()
```

De acuerdo con los resultados del análisis de Tukey:

- No hay diferencias significativas en la edad entre los métodos de ejecución de Electrocución, Firing Squad, Gas Chamber y Hanging en comparación con cada uno de los otros métodos. Esto significa que, en términos de la edad de los individuos ejecutados, no hay diferencias significativas entre estos métodos.

- Sin embargo, hay una diferencia significativa en la edad entre el método de ejecución de Electrocución y Lethal Injection. El meandiff es de 3.1395 con un valor p de 0.0008, lo que indica que la edad media de los individuos ejecutados mediante Lethal Injection es significativamente mayor que la edad media de los individuos ejecutados mediante Electrocución.

En resumen, parece que la edad de los individuos ejecutados varía significativamente entre los métodos de Electrocución y Lethal Injection, pero no difiere significativamente entre los otros métodos de ejecución.
