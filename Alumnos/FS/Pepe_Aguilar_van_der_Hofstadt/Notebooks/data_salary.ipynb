#Análisis de datos: Salario mundo data

##Introducción

Vamos a analizar un conjunto de datos sobre el salario de los data jobs.

Para ello se plantea la siguiente hipotesis "*Se percibe mayor salario al NO trabajar en remoto*".

Para comprobar si esta hipotesis es verdadera o falsa se utilizará un conjunto de datos de [kaggle](https://www.kaggle.com/datasets/lorenzovzquez/data-jobs-salaries).



```python
salary_df = pd.read_csv('salaries.csv')
salary_df.head(1)
```

En este conjunto de datos está la información sobre el puesto, el salario, el año, residencia, entre otros.

Para que el análisis sea lo más acertado posible hay que ajustar un poco los datos, por lo se modifican para que el entorno sea lo más parecido posible.

Para evitar el efecto que la inflación haya podido generar en los salarios, solo se utilian datos del año 2023.

```python
salary_df = salary_df.drop(salary_df.loc[salary_df['work_year'] < 2023].index)
```

Para el tema de la geografía se tendrán en cuenta solamente las personas cuyo salario provenga de una empresa estadounidense (ya que el 85% de las empresas del data set son estadounidenses)

```python
salary_df = salary_df.drop(salary_df.loc[salary_df['company_location'] != 'US'].index)
```

```python
alt.data_transformers.enable('default', max_rows=None)
```

```python
alt.Chart(salary_df).mark_bar().encode(
  x='employment_type',
  y='mean(salary_in_usd)',
  color='employment_type'
)
```
