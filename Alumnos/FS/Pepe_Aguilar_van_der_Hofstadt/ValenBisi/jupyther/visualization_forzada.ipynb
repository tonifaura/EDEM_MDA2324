# Proyecto ValenBisi

Vamos a visualizar el estado de una estaci√≥n de Valenbisi a lo largo del tiempo. 


```python
import json
from datetime import datetime
import matplotlib.pyplot as plt
```

## Estacion Ribera Plaza Ayuntamiento

```python
# Estacion Ribera_Plaza_Ayuntamiento
with open('Ribera_Plaza_Ayuntamiento.json', 'r') as file:
    json_data = json.load(file)

time = []
availables = []
frees = []

# obtener el estado de la estacion en cada punto
for entry in json_data:
    # extraccion de los valores
    time_point_str = entry["data"]["updated_at"].split(' ')[1]  # Extraer la parte de tiempo
    time_point = datetime.strptime(time_point_str, '%H:%M:%S').time()
    available = entry["data"]["available"]
    free = entry["data"]["free"]

    # guardado de los valores
    time.append(time_point)
    availables.append(available)
    frees.append(free)

# grafica
plt.figure(figsize=(5, 3))
timestamps = [datetime.combine(datetime.min, timestamp) for timestamp in time]  # Convertir a datetime.datetime
plt.plot(timestamps, availables, label='Bicis disponibles')
plt.plot(timestamps, frees, label='Huecos libres')

# Configurar etiquetas y leyenda
plt.xlabel('Tiempo')
plt.ylabel('n¬∫')
plt.title('Evoluci√≥n de bicicletas en la estacion Ribera Plaza Ayuntamiento')
plt.legend()

plt.xticks(rotation=45)
plt.yticks(range(int(min(min(availables), min(frees))), int(max(max(availables), max(frees)))+1))

# Mostrar el gr√°fico
plt.show()
```
