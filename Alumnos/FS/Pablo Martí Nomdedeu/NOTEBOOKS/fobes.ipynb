A continuación, vamos hacer un análisis a traves de gráficas para asegurar que las hipotesis que hemos establecido son las correctas, es decir, Los billonarios que logran su riqueza no tiene que ver nada la industria a la que se dediquen la edad o los estudios previos, es puramnente las necesidades de la sociedad.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/forbes_2640_billionaires.csv", error_bad_lines=False)
print(df)
```

Elegimos los 20 primers millonario para visualizar en que ambitos de industria nos movemos

```python
selected_columns = df.head(20)[['name', 'age', 'net_worth', 'industry']]
print(selected_columns)
```

Hacemos un grafico de barras para observar que:

```python
top_20_industries = df.head(20)['industry']

# Contar la frecuencia de cada industria
industry_counts = top_20_industries.value_counts()

# Crear un gráfico de barras
plt.figure(figsize=(12, 6))
industry_counts.plot(kind='bar', color='skyblue')
plt.title('Distribución de Millonarios por Industria (Top 20)')
plt.xlabel('Industria')
plt.ylabel('Número de Millonarios')
plt.xticks(rotation=45, ha='right')
plt.show()
```

La INDUSTRIA mas relevannte es la tecnológia y para ello seguimos comprobando que aunqu ela tecnología es la industria con mas millonarios, no es la que posee las mayores fortunas y aquellos que poseen las mayores fortunas son los individuos que han alcanzado una mayor edad

```python
selected_data = df.head(20)[['name', 'age', 'net_worth', 'industry']]

# Crear un gráfico de dispersión con colores por sector
plt.figure(figsize=(12, 8))
sns.scatterplot(x='age', y='net_worth', hue='industry', data=selected_data, palette='viridis', s=100)
plt.title('Relación entre Edad, Net Worth y Sector (Top 20)')
plt.xlabel('Edad')
plt.ylabel('Net Worth')
plt.legend(title='Industria', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Haremos un análisis de si, la educacón es la que ha ayudado a que estos usuario consigan alcanzar ese patrimonio tan extenso o a sido la edad

```python
df['Education_Total'] = df[['Bachelor', 'Master', 'Doctorate']].sum(axis=1)

# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='net_worth', y='Education_Total', hue='industry', data=df , s=100)
plt.title('Relación entre Net Worth, Educación Total e Industria')
plt.xlabel('Net Worth')
plt.ylabel('Educación Total')
plt.legend(title='Industria', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
